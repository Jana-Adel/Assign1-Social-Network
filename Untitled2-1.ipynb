```python
# 1) Imports & constants
import os
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from google.colab import files
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
```

```python
# 2) Upload dataset (Colab)
print("Please upload the SNAP facebook_combined file (facebook_combined or facebook_combined.txt)...")
uploaded = files.upload()
```

```python
# pick first uploaded filename
if len(uploaded) == 0:
    raise SystemExit("No file uploaded. Please upload facebook_combined file and re-run.")

uploaded_filename = list(uploaded.keys())[0]
print("Uploaded:", uploaded_filename)

DATA_DIR = Path("output")
DATA_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# 3) Read edgelist robustly (try several names / gz)
def try_read_edgelist(path):
    try:
        return nx.read_edgelist(path, nodetype=int)
    except Exception as e:
        raise e

# If the uploaded filename has extension .gz, networkx can read it if path provided.
# Use the local filename (it was saved by files.upload into working dir).
local_path = uploaded_filename
print("Reading edgelist from", local_path)
G = try_read_edgelist(local_path)
print("Graph loaded.")
print("Nodes:", G.number_of_nodes(), "Edges:", G.number_of_edges(), "Directed?", G.is_directed())
```

```python
# 4) Basic summary (NetworkX 3.4+ compatible)
print("\nGraph summary:")
print(" Nodes:", G.number_of_nodes())
print(" Edges:", G.number_of_edges())
print(" Density:", nx.density(G))
print(" Connected components:", nx.number_connected_components(G))
```

```python
# 5) Extract node-level features
nodes = list(G.nodes())
df = pd.DataFrame(index=nodes)

# degree
df['degree'] = pd.Series(dict(G.degree()))
# clustering
df['clustering'] = pd.Series(nx.clustering(G))
# triangles
df['triangles'] = pd.Series(nx.triangles(G))

# pagerank
try:
    df['pagerank'] = pd.Series(nx.pagerank(G))
except Exception as e:
    print("PageRank failed, falling back to np.zeros:", e)
    df['pagerank'] = 0.0

# betweenness (approx if graph large)
n_nodes = G.number_of_nodes()
if n_nodes > 3000:
    # approximate with k samples
    k = min(200, max(50, n_nodes // 20))
    print(f"Computing approximate betweenness with k={k}")
    df['betweenness'] = pd.Series(nx.betweenness_centrality(G, k=k, seed=RANDOM_SEED))
else:
    df['betweenness'] = pd.Series(nx.betweenness_centrality(G))

# average neighbor degree
df['avg_neighbor_deg'] = pd.Series(nx.average_neighbor_degree(G))

# core number
df['core_number'] = pd.Series(nx.core_number(G))

# Fill any NaN
df = df.fillna(0.0)
print("Feature matrix shape:", df.shape)
print("\nFeature matrix head:\n", df.head())
# Save raw features
df.to_csv(DATA_DIR / 'features_raw.csv')
# 6) Community detection (Louvain)
from community import community_louvain
partition = community_louvain.best_partition(G)
df['louvain'] = pd.Series(partition)
df.to_csv(DATA_DIR / 'features_with_communities.csv')
```

```python
# 7) Normalize selected features for ML
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_cols = ['degree', 'clustering', 'pagerank', 'betweenness', 'core_number', 'triangles', 'avg_neighbor_deg']
df_scaled = df.copy()
df_scaled[scaled_cols] = scaler.fit_transform(df_scaled[scaled_cols])
df_scaled.to_csv(DATA_DIR / 'features_scaled.csv')
print("Saved features_scaled.csv")
df_scaled.head()
```

```python
# 8) Create synthetic bot labels and create poisoned graphs
# We'll create LABELS dataframe (same index as df_scaled)
labels = pd.Series(0, index=df_scaled.index, name='label')  # 0 = benign

# Injection parameters
bot_fraction = 0.07  # 7% synthetic bots
budget_per_bot = 5
num_bots = max(1, int(len(df_scaled) * bot_fraction))
print(f"\nInjecting {num_bots} synthetic bots ({bot_fraction*100:.1f}%)")
# pick candidate nodes: low clustering, moderate degree (sort by clustering ascending, degree between pctiles)
candidates = df_scaled.sort_values(['clustering', 'degree']).index.tolist()
selected_bots = candidates[:num_bots]
labels.loc[selected_bots] = 1

# make a copy graph with injected manipulations (structural changes around bots)
G_bot = G.copy()
# pick hubs to connect bots to
hubs = [n for n,_ in sorted(G_bot.degree(), key=lambda x: x[1], reverse=True)][:200]

for v in selected_bots:
    neighs = list(G_bot.neighbors(v))
    # remove up to half the budget to lower clustering (if neighbors exist)
    removals = min(len(neighs), budget_per_bot // 2)
    for u in neighs[:removals]:
        if G_bot.has_edge(v, u):
            G_bot.remove_edge(v, u)
    # add edges to hubs
    adds = budget_per_bot - removals
    added = 0
    for h in hubs:
        if added >= adds:
            break
        if h == v:
            continue
        if not G_bot.has_edge(v, h):
            G_bot.add_edge(v, h)
            added += 1

# Save a GEXF for visualization (original + injected)
nx.write_gexf(G, DATA_DIR / 'graph_original.gexf')
nx.write_gexf(G_bot, DATA_DIR / 'graph_with_injected_bots.gexf')
```

```python
# 9) Structural Evasion Attack (on test-time) -> create another graph variant
# We'll simulate that attackers rewire selected_bots to appear more human
G_evasion = G_bot.copy()
for v in selected_bots:
    # strategy: remove some hub connections (that look synthetic) and instead connect to local communities
    # remove up to 2 hub edges if exist
    for h in list(G_evasion.neighbors(v))[:2]:
        if G_evasion.degree(h) > 50 and G_evasion.has_edge(v, h):
            G_evasion.remove_edge(v, h)
    # add edges between v and two neighbors-of-neighbors to increase local clustering
    neighbors = list(G_evasion.neighbors(v))
    for nb in neighbors[:2]:
        for nn in list(G_evasion.neighbors(nb))[:3]:
            if nn != v and not G_evasion.has_edge(v, nn):
                G_evasion.add_edge(v, nn)
nx.write_gexf(G_evasion, DATA_DIR / 'graph_after_evasion.gexf')
```

```python
# 10) Poisoning attack: add Sybil nodes connected to hubs to poison training
G_poison = G_bot.copy()
start_id = max(G_poison.nodes()) + 1
n_sybils = max(1, int(0.02 * G_poison.number_of_nodes()))  # 2% sybils
edges_per_syb = 10
hubs_list = [n for n,_ in sorted(G_poison.degree(), key=lambda x: x[1], reverse=True)][:500]
for i in range(n_sybils):
    new_id = start_id + i
    G_poison.add_node(new_id)
    # connect to random hubs
    targets = np.random.choice(hubs_list, size=min(edges_per_syb, len(hubs_list)), replace=False)
    for t in targets:
        G_poison.add_edge(new_id, int(t))
nx.write_gexf(G_poison, DATA_DIR / 'graph_poisoned.gexf')
```

```python
# 11) Recompute features for each scenario function
def compute_features_for_graph(G_obj):
    idx = list(G_obj.nodes())
    f = pd.DataFrame(index=idx)
    f['degree'] = pd.Series(dict(G_obj.degree()))
    f['clustering'] = pd.Series(nx.clustering(G_obj))
    f['triangles'] = pd.Series(nx.triangles(G_obj))
    f['avg_neighbor_deg'] = pd.Series(nx.average_neighbor_degree(G_obj))
    f['core_number'] = pd.Series(nx.core_number(G_obj))
    try:
        f['pagerank'] = pd.Series(nx.pagerank(G_obj))
    except:
        f['pagerank'] = 0.0
    # approximate betweenness if big
    if G_obj.number_of_nodes() > 3000:
        k = min(200, max(50, G_obj.number_of_nodes()//20))
        f['betweenness'] = pd.Series(nx.betweenness_centrality(G_obj, k=k, seed=RANDOM_SEED))
    else:
        f['betweenness'] = pd.Series(nx.betweenness_centrality(G_obj))
    f = f.fillna(0.0)
    return f

features_original = compute_features_for_graph(G)
features_with_injected = compute_features_for_graph(G_bot)
features_evasion = compute_features_for_graph(G_evasion)
features_poison = compute_features_for_graph(G_poison)
```

```python
# 12) Prepare ML dataset (we'll use features_with_injected as our dataset for training baseline; labels created earlier)
# Align labels to features_with_injected index; nodes absent => drop
nodes_common = features_with_injected.index.intersection(labels.index)
X = features_with_injected.loc[nodes_common, ['degree','clustering','pagerank','betweenness','core_number','triangles','avg_neighbor_deg']].fillna(0)
y = labels.loc[nodes_common].astype(int).values

print("\nTraining dataset size:", X.shape, "Positive (synthetic bots):", y.sum())
```

```python
# 13) Train baseline classifier on "clean" (here we treat G_bot as training structure with injected labels)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, precision_recall_curve, auc

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=RANDOM_SEED)
clf = RandomForestClassifier(n_estimators=200, random_state=RANDOM_SEED, class_weight='balanced')
clf.fit(X_train, y_train)

y_proba = clf.predict_proba(X_test)[:,1]
y_pred = clf.predict(X_test)
print("\nBaseline classifier report:\n", classification_report(y_test, y_pred))
print("Baseline ROC AUC:", roc_auc_score(y_test, y_proba))
```

```python
# 14) Evaluate on evasion graph (test-time structural evasion)
# compute features for nodes in test split using features_evasion
X_evasion = features_evasion.loc[X_test.index.intersection(features_evasion.index), X.columns].fillna(0)
# Align y_test for those indices
common_idx_evasion = X_test.index.intersection(X_evasion.index)
y_test_evasion = pd.Series(y_test, index=X_test.index).loc[common_idx_evasion].values
y_proba_evasion = clf.predict_proba(X_evasion.loc[common_idx_evasion])[:,1]
y_pred_evasion = clf.predict(X_evasion.loc[common_idx_evasion])
print("\nEvasion evaluation report:\n", classification_report(y_test_evasion, y_pred_evasion))
print("Evasion ROC AUC:", roc_auc_score(y_test_evasion, y_proba_evasion))
```

```python
# 15) Evaluate poisoned training scenario:
# retrain classifier on features_poison (simulate poisoned training data) then evaluate on clean test (features_original)
# For poisoning, we need labels for training - we'll use same labels (in practice poisoning can flip labels; here we simulate structural poisoning)
nodes_common_poison = features_poison.index.intersection(labels.index)
X_poison_train = features_poison.loc[nodes_common_poison, X.columns].fillna(0)
y_poison_train = labels.loc[nodes_common_poison].astype(int).values
# split train/test similarly (for fair comparison we use same random seed)
Xpt_train, Xpt_val, ypt_train, ypt_val = train_test_split(X_poison_train, y_poison_train, stratify=y_poison_train, test_size=0.3, random_state=RANDOM_SEED)

clf_poisoned = RandomForestClassifier(n_estimators=200, random_state=RANDOM_SEED, class_weight='balanced')
clf_poisoned.fit(Xpt_train, ypt_train)

# evaluate poisoned model on original clean features (nodes in our original test set intersection)
X_clean_test = features_original.loc[X_test.index.intersection(features_original.index), X.columns].fillna(0)
common_idx_poison_eval = X_clean_test.index
y_clean_test = pd.Series(y_test, index=X_test.index).loc[common_idx_poison_eval].values

y_proba_poison = clf_poisoned.predict_proba(X_clean_test)[:,1]
y_pred_poison = clf_poisoned.predict(X_clean_test)
print("\nPoisoned-training evaluation report (test on clean):\n", classification_report(y_clean_test, y_pred_poison))
print("Poisoned ROC AUC (test on clean):", roc_auc_score(y_clean_test, y_proba_poison))
```

```python
# 16) Save results summary
results = {
    'baseline': {
        'report': classification_report(y_test, y_pred, output_dict=True),
        'roc_auc': float(roc_auc_score(y_test, y_proba))
    },
    'evasion': {
        'report': classification_report(y_test_evasion, y_pred_evasion, output_dict=True),
        'roc_auc': float(roc_auc_score(y_test_evasion, y_proba_evasion))
    },
    'poisoned': {
        'report': classification_report(y_clean_test, y_pred_poison, output_dict=True),
        'roc_auc': float(roc_auc_score(y_clean_test, y_proba_poison))
    }
}

import json
with open(DATA_DIR / 'results_summary.json', 'w') as f:
    json.dump(results, f, indent=2)
print("Saved results summary to", DATA_DIR / 'results_summary.json')
```

```python
# 17) Visualize an example ego-net for one sample bot (before / evasion / poisoned)
sample_bot = selected_bots[0] if len(selected_bots)>0 else list(G.nodes())[0]
def plot_ego(G_obj, node, ax, title=''):
    ego = nx.ego_graph(G_obj, node, radius=2)
    pos = nx.spring_layout(ego, seed=RANDOM_SEED)
    node_cols = []
    for n in ego.nodes():
        if n == node:
            node_cols.append('red')
        elif n in selected_bots:
            node_cols.append('orange')
        else:
            node_cols.append('skyblue')
    nx.draw(ego, pos, node_color=node_cols, with_labels=False, node_size=80, ax=ax)
    ax.set_title(title)

fig, axs = plt.subplots(1,3, figsize=(18,6))
plot_ego(G_bot, sample_bot, axs[0], title='Injected bots (training graph)')
plot_ego(G_evasion, sample_bot, axs[1], title='After structural evasion (test)')
plot_ego(G_poison, sample_bot, axs[2], title='Poisoned graph (training)')
plt.tight_layout()
plt.savefig(DATA_DIR / 'ego_comparison.png', dpi=150)
plt.show()

print("\nAll outputs (CSV, GEXF, PNG, JSON) are in the folder:", DATA_DIR.resolve())
print("Download them from the Colab Files pane or use files.download if you want individual files.")
```
